## 8.
- Open the METABRIC data. Make a histogram of 'Ratio Therapy'.
- Let treatment, $T$ be distributed binomial with parameter $p$. Then the contribution to the likelihood for each patient $i$, with $y_i = 0$ for no radiation therapy and $y_i=1$ for radiation therapy, is 
$$
p^{y_i}(1-p)^{1-y_i}
$$
- Write out the likelihood.
- Maximize the likelihood with respect to $p$. What is the MLE, $\hat{p}$?
- Bootstrap the sampling density/distribution of $\hat{p}$.

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from scipy import stats

metabric = pd.read_csv("metabric.csv")

# Use 'Radio Therapy' column (binary: YES/NO)
radio_therapy = metabric['Radio Therapy'].dropna().str.upper().map({'YES': 1, 'NO': 0})
n = len(radio_therapy)

# Histogram 
radio_therapy.value_counts().sort_index().plot(kind='bar', color=['steelblue', 'orange'])
plt.xlabel("Radio Therapy (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("Histogram of 'Radio Therapy'")
plt.show()
```

Likelihood for binary observations $y_i\in\{0,1\}$:

$$
L(p) \;=\; \prod_{i=1}^n p^{y_i}(1-p)^{1-y_i}.
$$

Log-likelihood:

$$
\ell(p) \;=\; \sum_{i=1}^n \big( y_i \log p + (1-y_i)\log(1-p) \big).
$$

Differentiate and set to zero:

$$
\frac{d\ell}{dp} = \frac{\sum_{i=1}^n y_i}{p} - \frac{\sum_{i=1}^n (1-y_i)}{1-p} = 0
\quad\Rightarrow\quad
\hat{p} = \frac{1}{n}\sum_{i=1}^n y_i.
$$

So the MLE is the sample proportion: $\boxed{\hat p = \bar y}$.

```python
# MLE
p_hat = radio_therapy.mean()
print(f"n = {n},  p_hat = {p_hat:.4f}")

# Bootstrap sampling distribution of p_hat
B = 2000
boot_p = [np.mean(np.random.choice(radio_therapy, size=n, replace=True)) for _ in range(B)]
ci = np.percentile(boot_p, [2.5, 97.5])
print(f"Bootstrap 95% CI for p: ({ci[0]:.4f}, {ci[1]:.4f})")

# Bootstrap histogram
plt.hist(boot_p, bins=30, color='goldenrod', edgecolor='black')
plt.axvline(p_hat, color='red', linestyle='--', label='MLE p̂')
plt.title("Bootstrap sampling distribution of p̂")
plt.xlabel("p (proportion receiving Radio Therapy)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

The estimated probability of receiving radio therapy is p_hat = 0.6597 (65.97%). The bootstrap sampling distribution is narrow and approximately normal, yielding a 95% bootstrap CI of (0.6337, 0.6858). So it is fairly precise about the population proportion, with the histogram confirming a clear majority receiving radio therapy.

- Open the Ames house price data. Make a KDE of 'price'. Select an appropriate distribution for modeling it, and explain why you selected it. (Hint: You might want to take a common transformation of price.)
- Derive the density for this distribution.
- Write out the likelihood.
- Maximize the likelihood. What is the MLE?
- Plot the density/distribution for your fitted model and compare it to the KDE/ECDF. Criticize the fit.
- Bootstrap the sampling density/distribution of your parameters.

**Choose log-normal:** House prices are strictly positive and typically very right-skewed. Taking log(price) should produce a distribution much closer to symmetric / bell-shaped, so modeling log(price) with a Normal distribution is a natural, interpretable choice.

1. If $Y = \log(X)$ and $Y \sim \mathcal{N}(\mu, \sigma^2)$, then $X$ is lognormal and its density is:

$$
f_X(x) = \frac{1}{x\sigma\sqrt{2\pi}} 
\exp\!\left( -\frac{(\log x - \mu)^2}{2\sigma^2} \right), 
\quad x > 0
$$


2. Likelihood for $x_1, \ldots, x_n$ (lognormal, equivalently normal on the log scale):

$$
L(\mu, \sigma; x) 
= \prod_{i=1}^n f_X(x_i)
= \prod_{i=1}^n 
\left[ 
\frac{1}{x_i\sigma\sqrt{2\pi}} 
\exp\!\left( -\frac{(\log x_i - \mu)^2}{2\sigma^2} \right)
\right]
$$


On the log scale, with $y_i = \log x_i$, the likelihood becomes the normal likelihood:

$$
L(\mu, \sigma; y)
= (2\pi\sigma^2)^{-n/2} 
\exp\!\left( 
-\frac{\sum_{i=1}^n (y_i - \mu)^2}{2\sigma^2} 
\right)
$$


MLEs:

$$
\hat{\mu} = \frac{1}{n}\sum_{i=1}^n y_i,
\quad
\hat{\sigma}^2 = \frac{1}{n}\sum_{i=1}^n (y_i - \hat{\mu})^2
$$

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

RANDOM_STATE = 12345
N_BOOTSTRAP = 2000
np.random.seed(RANDOM_STATE)

ames = pd.read_csv('ames_prices.csv')

prices = ames['price'].dropna().astype(float)

# KDE plots (raw and log-transformed)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(prices, kde=True, ax=axes[0], stat="density", bins=60, color="skyblue")
axes[0].set_title("Price (raw) histogram + KDE")
axes[0].set_xlabel("Price")

log_prices = np.log(prices)
sns.histplot(log_prices, kde=True, ax=axes[1], stat="density", bins=60, color="lightgreen")
axes[1].set_title("Log(Price) histogram + KDE")
axes[1].set_xlabel("log(Price)")
plt.tight_layout()
plt.show()

# Fit lognormal via MLE
y = log_prices.values
n = len(y)
mu_mle = np.mean(y)
sigma2_mle = np.sum((y - mu_mle)**2) / n
sigma_mle = np.sqrt(sigma2_mle)

print("\nMLE estimates on log-scale:")
print(f"mu_hat = {mu_mle:.6f}")
print(f"sigma_hat = {sigma_mle:.6f}")

# Plot fitted density vs KDE
x_raw = np.linspace(prices.min(), prices.max(), 1000)
pdf_fitted_raw = stats.lognorm.pdf(x_raw, s=sigma_mle, scale=np.exp(mu_mle), loc=0)

plt.figure(figsize=(8, 5))
sns.kdeplot(prices, bw_method="scott", label="KDE (raw)", fill=False)
plt.plot(x_raw, pdf_fitted_raw, label="Fitted lognormal PDF", linewidth=2)
plt.title("Raw price: KDE vs fitted lognormal PDF")
plt.xlabel("Price")
plt.legend()
plt.show()

# ECDF vs fitted CDF
def ecdf(arr):
    xs = np.sort(arr)
    ys = np.arange(1, len(xs) + 1) / len(xs)
    return xs, ys

xs, ys = ecdf(prices)
cdf_fitted_raw = stats.lognorm.cdf(xs, s=sigma_mle, scale=np.exp(mu_mle), loc=0)
plt.figure(figsize=(8, 5))
plt.step(xs, ys, where='post', label='Empirical CDF')
plt.plot(xs, cdf_fitted_raw, label='Fitted Lognormal CDF', linewidth=2)
plt.title("ECDF vs Fitted Lognormal CDF")
plt.xlabel("Price")
plt.legend()
plt.show()
```

**KDE vs Fitted Lognormal PDF**

The lognormal model performs well overall, but with small deviations around the peak. The fitted lognormal curve follows the general right-skewed shape of the data quite well. The peak and the right tail roughly align, but the KDE’s peak is slightly sharper and higher than the fitted lognormal’s, meaning the model slightly underestimates the most common price range. The right tail (for higher prices) is fairly well captured, indicating that the lognormal model handles the skewness correctly.

**ECDF vs Fitted Lognormal CDF**

The near-perfect overlap means the lognormal is a good overall model for house prices on this dataset. Small deviations in the lower or upper tails may signal outliers or submarket heterogeneity (e.g., different housing segments or neighborhoods with distinct price scales).

```python
# Bootstrap sampling distribution for mu and sigma
B = N_BOOTSTRAP
mu_boot = np.empty(B)
sigma_boot = np.empty(B)

for b in range(B):
    sample_idx = np.random.randint(0, n, size=n)
    y_b = y[sample_idx]
    mu_b = np.mean(y_b)
    sigma_b = np.sqrt(np.sum((y_b - mu_b)**2) / n)
    mu_boot[b] = mu_b
    sigma_boot[b] = sigma_b

# Bootstrap confidence intervals
mu_ci = np.percentile(mu_boot, [2.5, 97.5])
sigma_ci = np.percentile(sigma_boot, [2.5, 97.5])

print("\nBootstrap 95% Confidence Intervals:")
print(f"mu_hat = {mu_mle:.6f},  CI = ({mu_ci[0]:.6f}, {mu_ci[1]:.6f})")
print(f"sigma_hat = {sigma_mle:.6f},  CI = ({sigma_ci[0]:.6f}, {sigma_ci[1]:.6f})")

# Plot bootstrap histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(mu_boot, bins=30, kde=True, ax=axes[0], color="orange")
axes[0].axvline(mu_mle, color='red', linestyle='--', label='MLE')
axes[0].set_title("Bootstrap distribution of mu (log-scale)")
axes[0].legend()

sns.histplot(sigma_boot, bins=30, kde=True, ax=axes[1], color="purple")
axes[1].axvline(sigma_mle, color='red', linestyle='--', label='MLE')
axes[1].set_title("Bootstrap distribution of sigma (log-scale)")
axes[1].legend()
plt.tight_layout()
plt.show()
```

## 10.
- Open the METABRIC data. Make a KDE of 'Overall Survival (Months)'.
- Let survival time, $T$ be distributed exponentially with parameter $\lambda$. Then its distribution is
$$
F(t) = 1 - e^{-\lambda t} = p[T\le t].
$$
- Derive the density for this distribution.
- Write out the likelihood.
- Maximize the likelihood with respect to $\lambda$. What is the MLE, $\hat{\lambda}$?
- Plot the density/distribution for your fitted model and compare it to the KDE/ECDF. Criticize the fit.
- Bootstrap the sampling density/distribution of $\hat{\lambda}$.

The density (PDF) is obtained by differentiating the CDF:

$$
f(t) = \frac{d}{dt}F(t) = \lambda e^{-\lambda t}, \quad t \ge 0.
$$

For observations $t_1, t_2, \ldots, t_n$, the likelihood is:

$$
L(\lambda) = \prod_{i=1}^n \lambda e^{-\lambda t_i}
= \lambda^n \exp\!\Big(-\lambda \sum_{i=1}^n t_i\Big).
$$

The log-likelihood is:

$$
\ell(\lambda) = \log L(\lambda) = n\log\lambda - \lambda \sum_{i=1}^n t_i.
$$

Differentiate with respect to $\lambda$ and set to zero:

$$
\frac{d\ell}{d\lambda} = \frac{n}{\lambda} - \sum_{i=1}^n t_i = 0
\quad \Rightarrow \quad
\hat{\lambda} = \frac{n}{\sum_{i=1}^n t_i}.
$$

MLE for $\lambda$ is the reciprocal of the sample mean:

$$
\boxed{\hat{\lambda} = \frac{1}{\bar{t}} = \frac{n}{\sum_{i=1}^n t_i}.}
$$

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

RANDOM_STATE = 12345
B = 2000  
np.random.seed(RANDOM_STATE)

metabric = pd.read_csv('metabric.csv')

survival_months = metabric["Overall Survival (Months)"].dropna().astype(float)
n = len(survival_months)

# MLE
lambda_hat = n / survival_months.sum()
print(f"MLE lambda_hat = {lambda_hat:.6g} (1/mean)")

# KDE
kde = stats.gaussian_kde(survival_months)
x_grid = np.linspace(0, survival_months.max()*1.05, 1000)

def pdf_exp(x, lam):
    return lam * np.exp(-lam * x)

def cdf_exp(x, lam):
    return 1 - np.exp(-lam * x)

# KDE and fitted exponential PDF
plt.figure(figsize=(8,5))
plt.plot(x_grid, kde(x_grid), label="KDE (density)", linewidth=1.5)
plt.plot(x_grid, pdf_exp(x_grid, lambda_hat), label=f"Fitted Exponential PDF (λ={lambda_hat:.4g})", linewidth=1.5)
plt.xlabel("Overall Survival (Months)")
plt.ylabel("Density")
plt.title("KDE of Survival Times and Fitted Exponential PDF")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# ECDF vs fitted CDF
sorted_t = np.sort(survival_months)
ecdf_y = np.arange(1, n+1) / n
plt.figure(figsize=(8,5))
plt.step(sorted_t, ecdf_y, where='post', label='Empirical CDF')
plt.plot(x_grid, cdf_exp(x_grid, lambda_hat), label='Fitted Exponential CDF', linewidth=1.5)
plt.xlabel("Overall Survival (Months)")
plt.ylabel("Cumulative proportion")
plt.title("ECDF vs Fitted Exponential CDF")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

```

**KDE vs fitted exponential**

The fitted exponential curve shows a smooth decline, while the KDE rises sharply at first, peaks near 50–100 months, then declines. Therefore, the exponential model underestimates short-term survival and overestimates long-term survival.

**ECDF comparison**

The fitted exponential CDF lies below the empirical ECDF up to about 200 months. This means the exponential model predicts patients to survive longer, on average, than the data actually show. After ~200 months, the exponential catches up, indicating the tails are somewhat consistent, but the early- to mid-survival range deviates strongly.

**Quantitative summary**: With lambda_hat = 1/t_bar = 0.007746 --> t_bar is approx 129 months. So the fitted mean survival time is roughly 129 months (≈10.7 years).

**Overall Critique:** exponential distribution is too restrictive for this dataset.


```python
# Bootstrap lambda_hat
lam_boot = np.empty(B)
for b in range(B):
    samp = np.random.choice(survival_months, size=n, replace=True)
    lam_boot[b] = len(samp) / samp.sum()

lam_mean = lam_boot.mean()
lam_std = lam_boot.std(ddof=1)
ci_low, ci_high = np.percentile(lam_boot, [2.5, 97.5])
print(f"Bootstrap mean(lambda) = {lam_mean:.6g}, std = {lam_std:.6g}")
print(f"95% bootstrap CI for lambda: ({ci_low:.6g}, {ci_high:.6g})")

plt.figure(figsize=(8,4))
plt.hist(lam_boot, bins=30, density=True, alpha=0.7)
x_l = np.linspace(lam_boot.min(), lam_boot.max(), 200)
plt.plot(x_l, stats.norm.pdf(x_l, loc=lam_mean, scale=lam_std), '--', label='Normal approx')
plt.axvline(lambda_hat, color='red', linestyle='--', label='MLE lambda_hat')
plt.xlabel('lambda (rate)')
plt.ylabel('Density')
plt.title('Bootstrap sampling distribution of lambda_hat')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## 11.
- Open the Ames house price data. Make a histogram of 'TotRms.AbvGrd', or total rooms above ground.
- We're going to model this using the Poisson Distribution. The probability that $y_i = k$ is given by
$$
pr[y_i = k ] = \frac{ \lambda^k e^{-\lambda} }{k!}
$$
- Write out the likelihood.
- Maximize the likelihood. What is the MLE? (Hint: The sample mean. As usual.)
- Plot the density/distribution for your fitted model and compare it to the KDE/ECDF. Criticize the fit.
- Bootstrap the sampling density/distribution of your parameter.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

RANDOM_STATE = 12345
N_BOOTSTRAP = 2000
np.random.seed(RANDOM_STATE)

ames2 = pd.read_csv('ames_prices.csv')

counts_raw = ames2['TotRms.AbvGrd'].dropna().astype(int)

plt.figure(figsize=(6,4))
sns.histplot(counts_raw, bins=20, color="C0")  
plt.title("Histogram of TotRms.AbvGrd")
plt.xlabel("TotRms.AbvGrd")
plt.ylabel("Frequency")
plt.show()
```

For $y_1, y_2, \ldots, y_n$, the likelihood function is:

$$
L(\lambda) = \prod_{i=1}^n \frac{\lambda^{y_i} e^{-\lambda}}{y_i!}.
$$

Taking the logarithm, the log-likelihood is:

$$
\ell(\lambda) = \sum_{i=1}^n \left( y_i \log \lambda - \lambda - \log(y_i!) \right).
$$

Differentiating with respect to $\lambda$ and setting to zero:

$$
\frac{d\ell}{d\lambda} 
= \frac{\sum_{i=1}^n y_i}{\lambda} - n = 0 
\quad \Rightarrow \quad 
\hat{\lambda} = \frac{1}{n}\sum_{i=1}^n y_i.
$$

MLE for $\lambda$ is simply the sample mean:

$$
\boxed{\hat{\lambda} = \bar{y} = \frac{1}{n}\sum_{i=1}^n y_i.}
$$

```python
# compute estimates
lam_hat = counts_raw.mean()
var_sample = counts_raw.var(ddof=1)
print(f"Poisson MLE (lambda_hat) = sample mean = {lam_hat:.4f}")

# Plot observed frequencies vs fitted Poisson frequencies
max_k = counts_raw.max()
k_vals = np.arange(0, max_k+1)
obs_counts = np.bincount(counts_raw, minlength=max_k+1)

# fitted pmf probabilities
n = len(counts_raw)
pmf_vals = stats.poisson.pmf(k_vals, mu=lam_hat)

# expected frequencies under fitted model
exp_counts = n * pmf_vals

plt.figure(figsize=(10,5))
plt.bar(k_vals - 0.2, obs_counts, width=0.4, label="Observed counts", color="C0")
plt.plot(k_vals + 0.2, exp_counts, 'o-', color="C3", label="Fitted Poisson expected counts", linewidth=2)
plt.xlabel("TotRms.AbvGrd (k)")
plt.ylabel("Frequency")
plt.title("Observed counts vs expected counts under fitted Poisson")
plt.legend()
plt.xticks(k_vals)
plt.show()

# ECDF comparison: empirical vs Poisson CDF
def ecdf(arr):
    xs = np.sort(arr)
    ys = np.arange(1, len(xs)+1)/len(xs)
    return xs, ys

xs, ys = ecdf(counts_raw)

# build fitted cdf at those xs
cdf_fitted = stats.poisson.cdf(xs, mu=lam_hat)
plt.figure(figsize=(8,5))
plt.step(xs, ys, where='post', label='Empirical CDF')
plt.plot(xs, cdf_fitted, label='Fitted Poisson CDF', linewidth=2, color='C3')
plt.xlabel("TotRms.AbvGrd")
plt.ylabel("Cumulative proportion")
plt.title("ECDF vs Fitted Poisson CDF")
plt.legend()
plt.show()

# 6) Goodness-of-fit
mean_sample = lam_hat
var_sample = counts_raw.var(ddof=1)
dispersion = var_sample / mean_sample
print(f"Mean = {mean_sample:.4f}, Variance = {var_sample:.4f}, Dispersion (Var/Mean) = {dispersion:.4f}")

# Bootstrap sampling distribution of lambda 
B = N_BOOTSTRAP
lam_boot = np.empty(B)
counts_arr = counts_raw.values
for b in range(B):
    samp = np.random.choice(counts_arr, size=n, replace=True)
    lam_boot[b] = samp.mean()

lam_ci = np.percentile(lam_boot, [2.5, 97.5])
print(f"Bootstrap mean(lambda) = {lam_boot.mean():.6f}")
print(f"95% bootstrap CI for lambda: ({lam_ci[0]:.4f}, {lam_ci[1]:.4f})")

plt.figure(figsize=(8,4))
sns.histplot(lam_boot, bins=40, kde=True)
plt.axvline(lam_hat, color='red', linestyle='--', label='MLE lambda')
plt.title("Bootstrap distribution of lambda (sample mean)")
plt.legend()
plt.show()
```

**Fit assessment.** The fitted Poisson (with lam_hat approx 6 from the sample mean) captures the overall right-skew and the broad shape of the room-count distribution. However, the fit shows a few discrepancies:

The Poisson PMF underpredicts the observed mass at the modal counts (around 5–7 rooms) and overpredicts probability at the low counts (0–3) and high counts (9-15).

The ECDF comparison shows small but noticeable deviations in the central region: the ECDF jumps more steeply near the mode than the fitted CDF, indicating that the data are more concentrated around the center than the Poisson predicts.

Taken together, these patterns suggest the Poisson assumption (variance equal to mean) may not hold exactly.
